```python
from cartographers.embeddings import EmbeddingExtractor
import numpy as np

print("=" * 50)
print("BERT Example")
print("=" * 50)

bert_extractor = EmbeddingExtractor(
    model_type='bert',
    model_name='bert-base-uncased'
)

# Extract from file
bert_embeddings = bert_extractor.extract_from_file('concepts_list.txt')
print(f"\nExtracted {len(bert_embeddings)} BERT embeddings")

# Save embeddings (optional)
np.savez('data/bert_embeddings.npz', **bert_embeddings)
```
